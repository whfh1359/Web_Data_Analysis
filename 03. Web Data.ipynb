### 1) find, select_one : 단일 선택
### 2) find_all, select : 다중 선택

```python
# soup.find_all("li", "on")  밑에는 똑같은 방법인데 select 사용
exchangeList = soup.select("li")
len(exchangeList) #모든 li에 해당되는게 가져와진다, id값으로 범위를 좁혀서 본다.
```

```python
# soup.find_all("li", "on")  밑에는 똑같은 방법인데 select 사용
# id : #
# class : > .
exchangeList = soup.select("#exchangeList > li")
len(exchangeList), exchangeList
```

```python
base_url = "https://finance.naver.com/"

title = exchangeList[0].select_one(".h_lst").text
exchange = exchangeList[0].select_one(".value").text
change = exchangeList[0].select_one(".change").text
updown = exchangeList[0].select_one(".head_info.point_up > .blind").text
link = base_url + exchangeList[0].select_one("a").get("href")
title, exchange, change, updown, link

```

# 4개 데이터 수집

```python
exchange_datas = []
base_url = "https://finance.naver.com/"

for item in exchangeList:
    data = {
        "title" : item.select_one(".h_lst").text,
        "exchange" : item.select_one(".value").text,
        "change" : item.select_one(".change").text,
        "updown" : item.select_one(".head_info.point_up > .blind").text,
        "link" : base_url + item.select_one("a").get("href"),

    }
    exchange_datas.append(data)
exchange_datas
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(exchange_datas)
df.to_excel("naverfinance.xlsx", encoding = "utf-8")
```

# 위키백과 문서 가져오기

```python
import urllib
from urllib.request import urlopen, Request

html = "https://ko.wikipedia.org/wiki/{search_words}"
req = Request(html.format(search_words = urllib.parse.quote("여명의 눈동자")))
response = urlopen(req)
soup = BeautifulSoup(response, "html.parser")
print(soup.prettify())

```

```python
n = 0
for each in soup.find_all("ul"):
    print("=> " + str(n) + "===================")
    print(each.get_text())
    n+=1
```

```python
#15번째 줄에 있다.
soup.find_all("ul")[15].text.strip().replace("\xa0","").replace("\n","")
```
